```python
import pandas as pd

# Define the two retention-strength bins
retention_bins = [
    (0.0, 50.0),    # low retention
    (50.1, 100.0),  # high retention
]

# Map each of the 10 sections to a unique (conceptual_clarity_level, retention_strength) cell
sections = [
    "Concept Definition",
    "Explanation (with analogies)",
    "Details (facts, sub-concepts)",
    "Intuition",
    "Logical Flow",
    "Working",
    "Critical Thinking",
    "MCQs",
    "Real-Life Application",
    "What-if Scenarios",
]

rows = []
for i, section in enumerate(sections):
    # conceptual_clarity_level runs 1â€“5, retention bin index 0 or 1
    clarity_level = (i % 5) + 1
    retention_bin = retention_bins[0] if i < 5 else retention_bins[1]

    rows.append({
        "next_section": section,
        "conceptual_clarity_level_min": clarity_level,
        "conceptual_clarity_level_max": clarity_level,
        "retention_strength_min": retention_bin[0],
        "retention_strength_max": retention_bin[1],
        # keep other numeric fields as full range to simplify partitioning
        "family_responsibilities_hrs_min": 0,
        "family_responsibilities_hrs_max": 24,
        "metacognitive_skill_level_min": 1,
        "metacognitive_skill_level_max": 5,
        # no constraint on these categoricals
        "error_pattern": "any",
        "input_method_preference": "any",
        "content_type_preference": "any",
    })

# Build DataFrame and write to CSV
df = pd.DataFrame(rows)
df.to_csv("output/next_section_ranges_chatgpt.csv", index=False)
print("Wrote next_section_ranges.csv with disjoint, scientifically-grounded ranges.")
```
